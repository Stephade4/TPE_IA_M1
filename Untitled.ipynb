```python
import numpy as np
import random

# ===============================
# 1️⃣ Définition de l'environnement 2D
# ===============================

class GridWorld:
    """
    Classe représentant l'environnement 2D.
    - L'agent se déplace dans une grille.
    - Certaines cases peuvent être des obstacles.
    - Une case finale (goal) donne une récompense positive.
    """

    def __init__(self, size=5, goal=(4, 4), obstacles=None):
        """
        size: taille de la grille (size x size)
        goal: position de la case objectif
        obstacles: liste de positions interdites (None = pas d’obstacles)
        """
        self.size = size
        self.goal = goal
        self.obstacles = obstacles if obstacles else []
        self.reset()

    def reset(self):
        """Remet l'environnement à l'état initial."""
        self.agent_pos = [0, 0]  # L’agent commence en haut à gauche (0,0)
        return tuple(self.agent_pos)

    def step(self, action):
        """
        Effectue une action et retourne :
        - next_state : nouvelle position
        - reward : récompense obtenue
        - done : si l’épisode est terminé
        """
        # Déplacements possibles
        moves = {
            0: (-1, 0),  # Haut
            1: (1, 0),   # Bas
            2: (0, -1),  # Gauche
            3: (0, 1)    # Droite
        }

        # Calcul de la nouvelle position
        move = moves[action]
        new_x = np.clip(self.agent_pos[0] + move[0], 0, self.size - 1)
        new_y = np.clip(self.agent_pos[1] + move[1], 0, self.size - 1)
        new_pos = (new_x, new_y)

        # Vérification des obstacles
        if new_pos in self.obstacles:
            reward = -10  # pénalité pour obstacle
            done = True
        elif new_pos == self.goal:
            reward = 10  # récompense pour atteindre la cible
            done = True
        else:
            reward = -1  # petite pénalité pour chaque pas
            done = False

        # Mise à jour de la position de l’agent
        self.agent_pos = list(new_pos)
        return new_pos, reward, done

```

```python
# ===============================
# 2️⃣ Création du modèle Q-Learning
# ===============================

class QLearningAgent:
    """
    Agent qui apprend grâce à l’algorithme Q-learning.
    """

    def __init__(self, env, alpha=0.1, gamma=0.9, epsilon=1.0, epsilon_decay=0.99, epsilon_min=0.10):
        """
        env: environnement (GridWorld)
        alpha: taux d’apprentissage (learning rate)
        gamma: facteur de récompense future
        epsilon: probabilité initiale d’exploration
        epsilon_decay: taux de réduction de epsilon à chaque épisode
        epsilon_min: valeur minimale de epsilon
        """
        self.env = env
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min

        # Création de la Q-table (états x actions)
        # Chaque case représente la "qualité" d’une action dans un état
        self.q_table = np.zeros((env.size, env.size, 4))  # 4 actions possibles

    def choose_action(self, state):
        """
        Choisit une action selon la politique epsilon-greedy :
        - Avec probabilité epsilon : action aléatoire (exploration)
        - Sinon : action ayant le meilleur Q-value (exploitation)
        """
        if random.uniform(0, 1) < self.epsilon:
            return random.randint(0, 3)  # exploration
        else:
            return np.argmax(self.q_table[state[0], state[1]])  # exploitation

    def learn(self, state, action, reward, next_state, done):
        """
        Met à jour la Q-table selon la formule du Q-learning :
        Q(s,a) ← Q(s,a) + α * [r + γ * max(Q(s’,a’)) − Q(s,a)]
        """
        old_value = self.q_table[state[0], state[1], action]
        next_max = np.max(self.q_table[next_state[0], next_state[1]])

        # Mise à jour de la valeur Q
        new_value = old_value + self.alpha * (reward + self.gamma * next_max - old_value)
        self.q_table[state[0], state[1], action] = new_value

        # Réduction progressive de epsilon (moins d’exploration au fil du temps)
        if done:
            self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

```

```python
# ===============================
# 3️⃣ Boucle d'entraînement
# ===============================

env = GridWorld(size=5, goal=(4, 4), obstacles=[(1,1), (2,1), (3,1), (3,2), (3,3), (3,4), (2,4), (1,4), (1,3)])
agent = QLearningAgent(env)

episodes = 1000  # Nombre total d’épisodes d’entraînement
rewards_per_episode = []

for episode in range(episodes):
    state = env.reset()
    total_reward = 0
    done = False

    while not done:
        action = agent.choose_action(state)
        next_state, reward, done = env.step(action)
        agent.learn(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward

    rewards_per_episode.append(total_reward)

print("✅ Entraînement terminé.")
print("Q-table finale :")
print(agent.q_table)
```

```python
import numpy as np
import gymnasium as gym

# === Création de l'environnement ===
env = gym.make("FrozenLake-v1", is_slippery=False)# render_mode="human")  # version déterministe

# === Initialisation Q-table ===
state_space = env.observation_space.n  # nombre d'états
action_space = env.action_space.n      # nombre d'actions
Q_table = np.zeros((state_space, action_space))

# === Hyperparamètres du Q-learning ===
alpha = 0.8       # taux d'apprentissage
gamma = 0.95      # facteur de récompense future
epsilon = 1.0     # taux d'exploration
epsilon_min = 0.05
epsilon_decay = 0.995
episodes = 3000

# === Entraînement ===
for episode in range(episodes):
    state, _ = env.reset()
    done = False

    while not done:
        # Choisir action (exploration/exploitation)
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q_table[state])

        # Exécuter action
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Mettre à jour Q-table
        best_next = np.max(Q_table[next_state])
        Q_table[state, action] += alpha * (reward + gamma * best_next - Q_table[state, action])

        state = next_state

    # Diminuer epsilon
    epsilon = max(epsilon_min, epsilon * epsilon_decay)

print("✅ Entraînement terminé !")
print("Q-table finale :\n", Q_table)
```

```python
class DeepQN(nn.Module):

    def __init__(self,input_dim, output_dim):
        super(DeepQN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64,64)
        self.fc3 = nn.Linear(64, output_dim)

    def sortie_neuronal(self,x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

```

```python
#creation de la memoire du modele
class memoire:
    def __init__(self, capacity):
        self.capacity = capacity
        self.souvenir=[]

    def add_remember (self,state, action, reward, next_state, done):
        #ajoute les experiences dans nos souvenir
        if len(self.souvenir) > self.capacity :
            self.souvenir.pop(0) #suppression du premier arriver afin de conserver l'effectif maximal de la memoire
        self.souvenir.append((state, action, reward, next_state, done))
    
    def choix_aleatoire(self, batch_size):
        return random.sample(self.souvenir, min(len(self.souvenir), batch_size))

    def __len__(self):
        return len(self.souvenir)
```

```python

# === Réseau de neurones Q ===
class QNetwork(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(QNetwork, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(state_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, action_dim)
        )

    def forward(self, x):
        return self.fc(x)

# === Environnement ===
env = gym.make("CartPole-v1")
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.n

# === Réseaux ===
q_net = QNetwork(state_dim, action_dim)
target_net = QNetwork(state_dim, action_dim)
target_net.load_state_dict(q_net.state_dict())  # copie initiale

optimizer = optim.Adam(q_net.parameters(), lr=0.001)
memory = deque(maxlen=50000)

# === Hyperparamètres ===
batch_size = 64
gamma = 0.99
epsilon = 1.0
epsilon_min = 0.05
epsilon_decay = 0.995
update_target_every = 1000

# === Entraînement ===
steps = 0
for episode in range(500):
    state, _ = env.reset()
    state = torch.FloatTensor(state)
    done = False

    while not done:
        # Choix action
        if random.random() < epsilon:
            action = env.action_space.sample()
        else:
            with torch.no_grad():
                action = torch.argmax(q_net(state)).item()

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        next_state = torch.FloatTensor(next_state)

        # Stocker la transition
        memory.append((state, action, reward, next_state, done))
        state = next_state

        # Apprentissage
        if len(memory) >= batch_size:
            batch = random.sample(memory, batch_size)
            s, a, r, s2, d = zip(*batch)
            s = torch.stack(s)
            s2 = torch.stack(s2)
            a = torch.tensor(a)
            r = torch.tensor(r)
            d = torch.tensor(d, dtype=torch.float32)

            q_values = q_net(s).gather(1, a.unsqueeze(1)).squeeze()
            with torch.no_grad():
                next_q = target_net(s2).max(1)[0]
                target = r + gamma * next_q * (1 - d)

            loss = nn.MSELoss()(q_values, target)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # Mise à jour réseau cible
        if steps % update_target_every == 0:
            target_net.load_state_dict(q_net.state_dict())

        steps += 1

    epsilon = max(epsilon_min, epsilon * epsilon_decay)

env.close()
print("✅ Entraînement terminé")
```
